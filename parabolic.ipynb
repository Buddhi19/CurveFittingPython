```python
#importing modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#Add data to the graph
x_values=[0,100,200,300,400,445,545,645,745]

y1_values=[0,0.94,1.69,2.33,2.57,2.66,2.38,1.72,0.9]

y2_values=[0,0.78,1.48,1.99,2.24,2.24,1.94,1.15,-0.21]
```

```python
#Vishualize given data
plt.scatter(x_values,y1_values)
plt.scatter(x_values,y2_values)
plt.show()
```

```python
#Curve Fitting
#Function we need to fit the graph

#Parabolic
def model_f1(x,a,b,c):
    return a*(x-b)**2 +c

#Linear
def model_f2(x,a,b):
    return a*x+b
```

```python
#use curve fit function to estimate values for a,b,c..

popt1, pcov1=curve_fit(model_f1,x_values,y1_values,p0=[-3,400,2.5])
print(*popt1)

popt2, pcov2=curve_fit(model_f1,x_values,y2_values,p0=[-3,400,2.5])
print(*popt2)
```

```python
#for graph 1
a1_opt, b1_opt, c1_opt = popt1
x_model = np.linspace(-100,1000,100)
y1_model = model_f1(x_model, a1_opt, b1_opt, c1_opt) 
 
plt.scatter(x_values, y1_values)
plt.plot(x_model, y1_model, color='r')

#for graph 2
a2_opt, b2_opt, c2_opt = popt2
x_model = np.linspace(-100,1000,100)
y2_model = model_f1(x_model, a2_opt, b2_opt, c2_opt) 
 
plt.scatter(x_values, y2_values)
plt.plot(x_model, y2_model, color='g')

##resize the graph
plt.xlim(-100,900)
plt.ylim(-0.5,3)
plt.show()
```
